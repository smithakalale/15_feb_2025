```python
#123
import numpy as np

list1 = [[10,20],[30,40],[50,60]]
val = np.array(list1)
print(val)

print(type(val))
print("dimension",val.ndim)
print("shape",val.shape)
print(val[1:])
print("transpose\n",val.T)

print("maximum",val.max())
print("minimum",val.min())
print("mean",val.mean())

sales=np.zeros(4)
print("zeros",sales)

sales=np.ones(5)
print("ones",sales)

sales = np.full(4,5)
print("full with 5s",sales)
print("flatten\n",val.flatten())


array = np.arange(10)
print(array)

print("mean =",np.average(array))
print("mean =",np.mean(array))
print("S.D=",np.std(array))
sd = np.sqrt(np.mean((array-np.mean(array))**2))
print("S.D=",sd)           
print("variance=",np.var(array))
print(np.median(array))

x = np.array([[0,3,4],[1,2,4],[3,4,5]])
print("shape",np.shape(x))
print("covariance=",np.cov(x))

x = np.array([1,3,5,7,8,9,10,15])
y = np.array([10,20,30,40,50,60,70,80])
print("covariance= \n",np.cov(x,y))
print("corelation = \n",np.corrcoef(x,y))

print("vstack= \n",np.vstack((x,y)))

print("roots= \n",np.roots(x))

print("dot product= ",np.dot(5,4))
vector_a = 2+3j
vector_b = 4+5j
print("dot = ",np.dot(vector_a,vector_b))

point1 = np.array((1,2,3))
point2 = np.array((1,1,1))

dist = np.linalg.norm(point1-point2)
print("Euclidean distance=",dist)
```
